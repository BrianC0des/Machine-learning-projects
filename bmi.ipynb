  Importing Dependecies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection

BMI Datasets

```python
#loading the bmi dataset to a pandas DataFrame
bmi_dataset = pd.read_csv('/content/bmi_realistic_dataset_10000_encoded.csv')
```

```python
#printing the first five rows
bmi_dataset.head()
```

```python
# number of rows and columns in this dataset
bmi_dataset.shape
```

```python
#getting the statstical measures of the dataset
bmi_dataset.describe()
```

```python
bmi_dataset['BMI_Category'].value_counts()
```

0 -> Underweight

1 -> Normal

2 -> Overweight

3 -> Obese

```python
bmi_dataset.groupby('BMI_Category').mean()
```

```python
x = bmi_dataset.drop(columns= 'BMI_Category', axis=1)
y = bmi_dataset['BMI_Category']
```

```python
print(x)
print(y)
```

Train Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Train the Model

```python
scaler = StandardScaler()
scaler.fit(x_train)
```

```python
classifier = svm.SVC(kernel='rbf', C=1.0, gamma='scale')
```

```python
# training the import vector machine classifier
classifier.fit(scaler.transform(x_train), y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy score on the training
x_train_prediction = classifier.predict(scaler.transform(x_train))
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy score of training data: ', training_data_accuracy)
```

```python
x_test_prediction = classifier.predict(scaler.transform(x_test))
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy score of the data: ',test_data_accuracy  )
```
